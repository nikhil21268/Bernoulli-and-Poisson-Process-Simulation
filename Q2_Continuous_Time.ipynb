This code is same as Q2.ipynb - except that, it uses continuous time values (by making use of np.linspace() function)

This, in turn, makes the function behave similar to a continuous time RV.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define parameters
lambda_ = 5  # Arrival rate (average arrivals per unit time)
t = np.linspace(0, 10, 10000000)  # Define a range of continuous time values
```

```python
# Simulate number of arrivals for each time point in t
arrivals = np.random.poisson(lambda_, len(t))
```

```python
# Plot the histogram
plt.hist(arrivals, bins=np.arange(0, np.max(arrivals)+2)-0.5, density=True, edgecolor='black', alpha=0.7)
plt.title(f'Density of Number of Arrivals until time t for Î» = {lambda_}')
plt.xlabel('Number of Arrivals')
plt.ylabel('Density')
plt.show()
```

```python
# Define parameters
lambda_15 = 15  # Arrival rate (average arrivals per unit time)
t = np.linspace(0, 10, 10000000)  # Define a range of continuous time values

# Simulate number of arrivals for each time point in t
arrivals = np.random.poisson(lambda_15, len(t))

# Plot the histogram
plt.hist(arrivals, bins=np.arange(0, np.max(arrivals)+2)-0.5, density=True, edgecolor='black', alpha=0.7)
plt.title(f'Density of Number of Arrivals until time t for Î» = {lambda_15}')
plt.xlabel('Number of Arrivals')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
import math

# Define parameters
lambda_ = 5  # Arrival rate

# Define the number of simulations
num_simulations = 1000

# Generate inter-arrival times for each simulation
inter_arrival_times = []

for _ in range(num_simulations):
    inter_arrival_time = -1/lambda_ * math.log(random.random())
    inter_arrival_times.append(inter_arrival_time)

# Plot histogram
plt.hist(inter_arrival_times, bins=30, density=True, edgecolor='black', alpha=0.7)
plt.title('Density of First Inter-Arrival Time')
plt.xlabel('Inter-Arrival Time')
plt.ylabel('Density')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
import math

# Define parameters
lambda_ = 5  # Arrival rate

# Define the number of simulations
num_simulations = 1000

# Define the number of inter-arrival times to generate
num_inter_arrival_times = 10000

# Generate inter-arrival times for each simulation
inter_arrival_times = []

for _ in range(num_simulations):
    inter_arrival_time = [-1/lambda_ * math.log(random.random()) for _ in range(num_inter_arrival_times)]

    inter_arrival_times.append(inter_arrival_time)

# Plot histograms for each simulation
for i in range(num_simulations):
    plt.hist(inter_arrival_times[i], bins=30, density=True, alpha=0.5)

plt.title('Density of First Inter-Arrival Time')
plt.xlabel('Inter-Arrival Time')
plt.ylabel('Density')
plt.show()
```
