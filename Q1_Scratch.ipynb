```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
num_simulations = 1000
num_trials = 20

# Function to simulate Bernoulli trials
def simulate_bernoulli(p, num_trials, num_simulations):
    results = np.zeros((num_simulations, num_trials), dtype=int)
    for i in range(num_simulations):
        results[i, :] = np.random.choice([0, 1], num_trials, p=[1-p, p])
    return np.sum(results, axis=1)

# (a) Simulation for p = 0.8
p_a = 0.8
results_a = simulate_bernoulli(p_a, num_trials, num_simulations)

# (b) Simulation for p = 0.5
p_b = 0.5
results_b = simulate_bernoulli(p_b, num_trials, num_simulations)

# Plot histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(results_a, bins=np.arange(0, num_trials+2)-0.5, edgecolor='black', alpha=0.7)
plt.title(f'Simulation for p = {p_a}')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(results_b, bins=np.arange(0, num_trials+2)-0.5, edgecolor='black', alpha=0.7)
plt.title(f'Simulation for p = {p_b}')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
