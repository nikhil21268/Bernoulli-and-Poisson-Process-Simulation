(a) Simulate the density of number of arrivals until time t for Î» = 5.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define parameters
lambda_ = 5
t = 10  # Time interval
```

```python

# Generate Poisson random variable
arrivals = np.random.poisson(lambda_ * t, 10000)
```

```python
# Plot the histogram
plt.hist(arrivals, bins=np.arange(0, np.max(arrivals)+2)-0.5, density=True, edgecolor='black', alpha=0.7)
plt.title(f'Density of Number of Arrivals until time t for Î» = {lambda_}')
plt.xlabel('Number of Arrivals')
plt.ylabel('Density')
plt.show()
```

```python
# Calculate the mean value
mean_arrivals = np.mean(arrivals)
print(f'Mean number of arrivals: {mean_arrivals}')
```

This code simulates the density of the number of arrivals in the time interval (0, t] for a Poisson process with Î» = 5. It generates 10,000 samples and plots a histogram showing the density. The mean number of arrivals is calculated and printed.

The mean comes out to be, Î» * t = 50 (approximately).

Here, t <- time interval (viz. 10 in my implementation - can be adjusted in the code, as per requirement)

Now, the below code does the exact same thing as part (c), except that it uses 'num_simulations' as a parameter too.

This plots 1000 simulations of the exponential RV on the graph, & the result is that, the graph appears more smoother, & the differences for each simulation is clearly visible.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
lambda_ = 5  # Arrival rate

# Define the number of simulations
num_simulations = 1000

# Generate inter-arrival times for each simulation
inter_arrival_times = np.random.exponential(scale=1/lambda_, size=(num_simulations, 10000))

# Plot histograms for each simulation
for i in range(num_simulations):
    plt.hist(inter_arrival_times[i], bins=30, density=True, alpha=0.5)

plt.title('Density of First Inter-Arrival Time')
plt.xlabel('Inter-Arrival Time')
plt.ylabel('Density')
plt.show()
```
