This file (i.e. Q2_Scratch.ipynb) essentially functions the same way as Q2.ipynb - however, here, we're implementing the Poisson distribution ourselves - without making use of the built-in functions for Poisson RV, of the NumPy library.

```python
import matplotlib.pyplot as plt
import random

# Define parameters
lambda_ = 5  # Arrival rate (average arrivals per hour)
t = 10  # Time interval (in hours)

# Number of simulations
num_simulations = 10000

# Simulate number of arrivals for each simulation
arrivals = [sum(random.expovariate(1/lambda_) for _ in range(int(t))) for _ in range(num_simulations)]

# Plot the number of arrivals as a function of time
plt.hist(arrivals, bins=range(int(max(arrivals))+2), density=True, edgecolor='black', alpha=0.7)
plt.title(f'Number of Arrivals over Time (Î» = {lambda_})')
plt.xlabel('Number of Arrivals')
plt.ylabel('Density')
plt.show()

# Calculate the mean value
mean_arrivals = sum(arrivals) / num_simulations
print(f'Mean number of arrivals: {mean_arrivals}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
lambda_ = 5  # Arrival rate

# Define the number of simulations
num_simulations = 1000

# Generate inter-arrival times for each simulation
inter_arrival_times = np.random.exponential(scale=1/lambda_, size=(num_simulations, 10000))

# Plot histograms for each simulation
for i in range(num_simulations):
    plt.hist(inter_arrival_times[i], bins=30, density=True, alpha=0.5)

plt.title('Density of First Inter-Arrival Time')
plt.xlabel('Inter-Arrival Time')
plt.ylabel('Density')
plt.show()
```
